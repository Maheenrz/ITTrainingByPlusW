```python
#Task_01

tasks = []  

# Add Tasks  
def add_task(task):  
    tasks.append(task)  

# Remove Tasks  
def remove_task(task):  
    if task in tasks: 
        print("Removing: ", task) 
        tasks.remove(task)  

# Function to View Tasks  
def view_tasks():  
    for task in tasks: 
        print(task)  


add_task("Finish homework")  
add_task("Buy groceries")  
add_task("Buy clothes")  
view_tasks()  
remove_task("Buy groceries")  
view_tasks()
```

```python
 # Task 2
def find_max_min(lst):  
    # Initialize max_val and min_val  
    max_val = lst[0]  
    min_val = lst[0]  

 
    for num in lst:  
        # Comparing min,max 
        if num > max_val:  
            max_val = num  
        if num < min_val: 
            min_val = num  


    return max_val, min_val  


numbers = [3, 9, 2, 8, 1]  
max_val, min_val = find_max_min(numbers)  
print(f"Max: {max_val}, Min: {min_val}")  
```

```python
  # Task 3
phonebook = {}  

# Add Contacts  
def add_contact(name, number):  
    phonebook[name] = number  

 
def get_number(name):  
    return phonebook.get(name, "Contact not found")  

# Example Usage  
add_contact("Alice", "1234567890")  
add_contact("Bob", "9876543210")  
print(get_number("Alice"))  
```

```python
 # Task 4
  
inventory = {}  

# Add Items  
def add_item(item, quantity):  
 
    if item in inventory:  
        inventory[item] += quantity  
    else:  
        inventory[item] = quantity  

# Remove Items  
def remove_item(item, quantity):  
 
    if item in inventory and inventory[item] >= quantity:  
        inventory[item] -= quantity  

        if inventory[item] == 0:  
            del inventory[item]  
    else:  
        print(f"Insufficient quantity of {item}.")  

# View Inventory  
def view_inventory():  
    for item, quantity in inventory.items():  
        print(f"{item}: {quantity}")  

# Example Usage  
add_item("Apple", 10)  
add_item("Banana", 5)  
remove_item("Apple", 3)  
view_inventory()  
```

```python
# Task 5
# Deposit   
def deposit(balance, amount):  
    balance += amount  
    print(f"Deposited: {amount}. New balance: {balance}")  
    return balance   

# Withdraw   
def withdraw(balance, amount):  
    if amount <= balance:  
        balance -= amount  
        print(f"Withdrew: {amount}. New balance: {balance}")  
    else:  
        print("Insufficient funds")  
    return balance    

#  Check Balance  
def check_balance(balance):  
    print(f"Current balance: {balance}")    

# Example Usage  
balance = 1000  
balance = deposit(balance, 500)  
balance = withdraw(balance, 200) 
check_balance(balance) 
```
